```python
import networkx as nx
import networkx.algorithms.community as nx_comm
import matplotlib.pyplot as plt
import numpy as np
import scipy as scp
import pandas as pd
import node2vec as n2v
import karateclub as kc
import random
import stellargraph as sp
from gem.embedding.gf import GraphFactorization
```

```python
def draw_graph(G, pos_nodes, node_names={}, node_size=50, plot_weight=False):
    nx.draw(G, pos_nodes, with_labels=False, node_size=node_size, edge_color='gray', arrowsize=30)
    
    pos_attrs = {}
    for node, coords in pos_nodes.items():
        pos_attrs[node] = (coords[0], coords[1] + 0.08)
        
    nx.draw_networkx_labels(G, pos_attrs, font_family='serif', font_size=20)
    
    
    if plot_weight:
        pos_attrs = {}
        for node, coords in pos_nodes.items():
            pos_attrs[node] = (coords[0], coords[1] + 0.08)
        
        nx.draw_networkx_labels(G, pos_attrs, font_family='serif', font_size=20)
        edge_labels=dict([((a,b,),d["weight"]) for a,b,d in G.edges(data=True)])
        nx.draw_networkx_edge_labels(G, pos_nodes, edge_labels=edge_labels)
    
    plt.axis('off')
    axis = plt.gca()
    axis.set_xlim([1.2*x for x in axis.get_xlim()])
    axis.set_ylim([1.2*y for y in axis.get_ylim()])
```

<b>Observe</b> Graphs can be analyzed at different levels (node, edge, graph level)

- Node Level (Given a graph $G = (V, E)$ classify each $ v \in V$ into right classes 
- Edge Level: Given a graph $G = (V, E)$, classify edge into right class (i.e. $<ei, yi>$) where $ei$ is an edge of graph $G$ and $yi$ is where it belongs, link prediction
- Graph level: given $m$ different graphs, put graph into right class

<b>Observe</b> Need to create features engineering of graphs, extract certain structural attributes.  THis can include standard, but also kernal functions

<b>Observe</b> Extensive work create algorihms of learning a good representation, that reflect topology - representation learning learning a good representation of a graph



<b>Definition</b>: Representation Learning (network embedding) is the task that aaims to learn a mapping function from a ddiscrete graph to continous domain function, low dimensional representation

$$ f: g \rightarrow \mathbb{R}^n $$

Note this can be at multiple levls such as $f:V \rightarrow \mathbb{R}^n$ and $f:E \rightarrow \mathbb{R}^n$

<b>Example</b>

```python
G = nx.barbell_graph(m1 = 7, m2 = 4)
draw_graph(G, nx.spring_layout(G))
```

```python
# need info on 
node2vec = n2v.Node2Vec(G, dimensions=2)
model = node2vec.fit(window=10)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

for x in G.nodes():
    
    v = model.wv.get_vector(str(x))
    ax.scatter(v[0],v[1], s=1000)
    ax.annotate(str(x), (v[0],v[1]), fontsize=12)
```

<b>Example</b>: 

Edge to vector example

```python
edges_embs = n2v.edges.HadamardEmbedder(keyed_vectors=model.wv)
```

```python
fig, ax = plt.subplots(figsize=(10,10))

for x in G.edges():
    
    v = edges_embs[(str(x[0]), str(x[1]))]
    ax.scatter(v[0],v[1], s=1000)
    ax.annotate(str(x), (v[0],v[1]), fontsize=16)
```

```python


n_graphs = 20

def generate_radom():
    # this will create a bunch of graphs
    n = random.randint(6, 20)
    k = random.randint(5, n)
    p = random.uniform(0, 1)
    return nx.watts_strogatz_graph(n,k,p), [n,k,p]

Gs = [generate_radom() for x in range(n_graphs)]

model = kc.Graph2Vec(dimensions=2, wl_iterations=10)
model.fit([x[0] for x in Gs])
embeddings = model.get_embedding()

fig, ax = plt.subplots(figsize=(10,10))

for i,vec in enumerate(embeddings):
    
    ax.scatter(vec[0],vec[1], s=1000)
    ax.annotate(str(i), (vec[0],vec[1]), fontsize=40)
```

### Embedding

<b>Observe</b>methods classified into transductive and inductive, depending on update procedure when new samples are added

- transductive - new nodes are provided lead to retraining the model
- inductive - models expected to generalise to new nodes, edges, graphs not observed in training 

<b>Ref</b> https://arxiv.org/pdf/2005.03675.pdf, Machien Learning on Graphs: A model and Comprehensive Taxonomy

<b>Definition</b> An <b>Encoder</b> maps an input to an embedding space

<b>Definition</b> A <b>Decoder</b> decodes structural information about graph from embedding

<b>Observe</b> The main reason is compression

<b>Observe</b> A standard way is embed using distance/proxmiity

<b>Observe</b>: there are 4 types of embedding algorithms: 

1. shallow embedding methods (i.e. Node2Vec, Edge2Vec, Graph2Vec). Limited to only seen data, can't infer. Basic mapping of input graphs into vectors - unsupervised / supervised 

2. Graph embedding methods - does basic, also learns a more gen functiion $f(G)$ - unsepervised 

3. Neighborhood aggregation methods - extract embeddings at graph level where nodes are labled with preoperties, also use $f(G)$ to get overall, and can also use some node properties  - supervised and unsupervised 

4. Graph regularisation methods - graph is not imput, but learn interactions of features, graph constructued by learning about feature similiarity. These agorithms learn $f(X)$ which maps features $X$ to embedding vector - semi supervised / supervised



<b>Observe</b>: All models use loss functinons in supervised to ascertain diff between pred and target, or evaulate siilirity between graph and graph created from ENC + DEC
