Some things to note from this video: 

* It is possible to think of a neuron in a couple of interesting ways. The first is as a 2-state machine that observes a given, particular input and based on this input, will go into an 'on' or 'off' state. The second is a machine that observes the given state of many other neuron's and based on this state will go into and 'on' or 'off' state. Whichever we think of this (and they are very similiar), we are dealing with a 2-state machine that observes and is being affected by its environment and reacts by changing (or not) current state. 

* At any given time (and note we have not really defined what "any given time" really means), a number of neurons in the neocortex will be in a state of off or on. If you took a snapshot of this activity, you could imagine that you might see something like one of the SDRs we visualised in the last notebook: a big grid in which you can visualise each of the squares as being a neuron, and where most are in the 'off' state (the squares that are white). And sparsity is in play to: at any time, only a small number (the squares that are colored) that are in an 'on' state. In this way, the SDR could be considered as as a record of many neuron states at a given time.  

* Note however, that this characterisation does not really do enough justice to the importance of the SDR data structure. The SDR is a data structure that can help us understand how information moves through all parts of the heirachy, allowing neurons to recieve both inputs and context and it can be interpreted in different ways.

* If we accept that SDRs can represent some information (a specific state space where each of the states is set to 0 or 1) it becomes critical to understand when differnt spaces could be regarded as the same, or regarded as similiar. We can look at this information in this different ways. Last time we took the view that, if I have 2 SDRs (say SDR1 and SDR2) there is going to be some likliehood that some active bits in SDR1 and SDR2 may share the same positions, and I can call the count of these the <b>overlap score</b>



- overlap set is the total number of SDRs that have the same n and a specific w with a specific number of bits of overlap

- ho many ways can these specific bits fit into these specific spots

But I could take a much broader view of the problem. Let's say I consider every possible SDR given some size and population and then compare every one of those to every other possible SDR given the some size and population? How many of these SDR combinations would have an overlap (i.e. have at least some number of common active bits) in some way to my original SDR for some some threshold that set? And if I set some arbitrary score (say an overlap score of 5), I might then say that any overlap of 5 or above is a match. 

This is a nice way to think about things, because it tells us both about the enormous capacity of our space, and how effective our space can be representing semantic similiarity. 

To understand this further, let's define the <b>Overlap Set</b> of an SDR:

$$  |\Omega_x(n, w, b| = \binom{w_0}{b} \times \binom{n - w_0}{w_1 - b}  $$

So what do the variables in this formula mean? 

* $n$ is the size of the the input space for any SDR that is being created
* $w_0$ is the population (number of active bits) in my original SDR, the one which I will use to compare to all other possible SDRs
* $w_1$ is the population of all other SDRs that can be be ceated. 
* $b$ is the number that I am free to set, and represents the least amount of overlap bits that I require to regard my SDR to be considered a match. Think of this as a matching threshold.

The best way to think of this is that, $\binom{w_0}{b}$ tells me all the ways I could arrange active bits in the 'on' space that meet my threshold, and  $\binom{n - w_0}{w_1 - b} $ tells me all the ways I can arrange the the remaining bits in the the rest of the space. I can then multiply this all together to get the total number of ways bits could be arranged.

left side - on bit space, right side off bit space
40 on bits, can only fit in 40 spaces

Let's create this function in Python:

```python
def computeOverlapSetCardinality(n, w0, w1, b, provide_summary = True):
    on_bit_space = sp.binomial(w0, b)
    off_bit_space = sp.binomial(n - w0, w1 - b)
    overlap_set = on_bit_space * off_bit_space
    if provide_summary:
        print( "The total number of number of ways to arrange", str(b), "active bits for all SDRs each with", str(w0),  
              "active bits in an input space size of" , str(n), ":\n", on_bit_space)
        print("For all bits in the space that are not active-bit, being a count of", 
              str(n - w0), ", the number of ways to arrange", str(w1 - b), "off-bits:\n", off_bit_space)
        print("Number of SDRs that can be considered a match with these constraints:\n", overlap_set)
    return(overlap_set)
```

Another thing to consider So if I get 30 bits out of my 40 what are the odds that I don't have the 40? We can compute this by dividing the overlap set cardinatoity with how unique the original SDR is. Lets create a Python function for this:

```python
def compute_likelihood_of_false_postive(n, w0, w1, b):
    return(float(compute_overlap_set_cardinality(n, w0, w1, b) / sp.binomial(n, w0)))
```

```python
compute_likelihood_of_false_postive(600, 40, 40, 30)
```

So this says if I return and SDR that has 30 of the 40 bits, this is the likelihood at it not my original SDR.

```python
def create_randominsed_sdr(x_limit, y_limit, population, one_dimensional = False):
    sdr = set()
    
    if one_dimensional:
        while len(sdr) < population:
            sdr.add(np.random.randint(low=1, high=x_limit))
    
    else:                
        while len(sdr) < population:
            sdr.add((np.random.randint(low=1, high=x_limit),np.random.randint(low=1, high=y_limit)))
    
                
    return(list(sdr))

def create_axis_for_sdr(ax, x_limit, y_limit, population, label):
    
    label_add = np.round((population / (x_limit * y_limit)) * 100, 2)
    label = label + ' (Sparsity: {}%)'.format(label_add)
    
    ax.set_xlabel(label)
    ax.set_xticks(range(int(x_limit)))
    ax.set_yticks(range(int(y_limit)))
    [ax.xaxis.get_major_ticks()[i].tick1line.set_color("white") for i in range(int(x_limit))]
    [ax.yaxis.get_major_ticks()[i].tick1line.set_color("white") for i in range(int(y_limit))]
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.grid(color='k', linestyle='-', linewidth=.5)
    
    return(ax)

def compute_union_and_overlap(SDR1_on_bits, SDR2_on_bits):
    union = list(set(SDR1_on_bits).union(SDR2_on_bits))
    overlap = list(set(SDR1_on_bits).intersection(SDR2_on_bits))
    
    return({"union": union, "overlap": overlap})

def create_and_compare_sdrs_over_multiple_iterations(iterations, sdr_size, population):
    # make this return averages
    sdr_unions_for_comparison = []
    sdr_overlaps_for_comparison = []
    for x in range(iterations):
        SDR1 = create_randominsed_sdr(np.sqrt(sdr_size), np.sqrt(sdr_size), population)
        SDR2 =create_randominsed_sdr(np.sqrt(sdr_size), np.sqrt(sdr_size), population)
        sdr_comparison = compute_union_and_overlap(SDR1, SDR2)
        sdr_unions_for_comparison.append(len(sdr_comparison['union']))
        sdr_overlaps_for_comparison.append(len(sdr_comparison['overlap']))
    print("Average union: ", str(sum(sdr_unions_for_comparison) / len(sdr_unions_for_comparison)))
    print("Average overlap: ", str(sum(sdr_overlaps_for_comparison) / len(sdr_overlaps_for_comparison)))
    return({"union_comparison": sdr_unions_for_comparison, "overlap_comparison": sdr_overlaps_for_comparison})
    
def randomly_flip_percentage_of_bits(SDR, array_size, percentage_to_flip):
    to_remove = int(percentage_to_flip * len(SDR))
    randomised_SDR = np.random.shuffle(SDR)
    del SDR[0:to_remove]
    noise = create_randominsed_sdr(np.sqrt(array_size), np.sqrt(array_size), to_remove)
    sdr_with_noise = noise + SDR
    return(sdr_with_noise)

```
