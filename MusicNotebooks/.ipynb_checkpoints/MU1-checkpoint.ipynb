# Transposing Instruments




There is a complicated history around why we have transposing instruments, but it really just comes down to the fact it tends to make passages easier to read, and allows instrumentalists to move between instruments. A typical example, clarinetist can apply the same fingering across all members of the same family. 

There are two issues to content with when it comes to transposing instruments score reading. The first is that when you encounter the notes of a transposing instrument written on a score, you will to transpose these notes to find out which concert pitch is being played. For example, a C4-Note played on a Bb Clarinet will sound a M2 lower (Bb3). Alternatively, playing a C5-Note in an Alto Flute (a transposing instrument in the key of G) will result in a concert note that is P4 lower, a G4-Note. To read scores effectively, you need to be able to quickly transpose notes to understand which concert pitch they are. 

The other thing you will notice about transposing instruments is that they appear in a different key on a score. This can be difficult to understand, and the logic is best illustrated with an example: suppose I want the notes G-F-E in C Major played. If these are read by a Bb clarinetist, they will result in F-Eb-D in concert pitches. So I will need to provide the clarinetist with the notes A-G-F#. These are in the key of C Major. If I write a passage in D Major, The transposition will mean that all written notes need to be transposed by a given interval (down a M2) and I will adjust the key signature to minimise the sharps and flats.

- <b>Transposing flats:</b> <br/> 
All flats become flats a minor 3rd lower, Eb Ab, Bb which become naturals a minor 3rd lower
