To get started, let's bring back our Python function to create a Pascal Matrix. 


```python
def compute_pascal_array(dimension_choice):
    all_entries = []
    for i in range(dimension_choice):
        row_entries = []
        for j in range(dimension_choice):
            row_entries.append( sp.binomial(i, j))
        all_entries.append(row_entries)
    
    return(sp.Matrix(all_entries))
    
```

Start by recalling the idea of telescoping series: 

$  \sum_{m=1}^n(m^k-(m-1)^k)$<br/>
$=(1^k-0^k)+(2^k-1^k)+\cdots+(n^k-(n-1)^k)$ <br/>
$ =n^k-0^k=n^k.  $

The arithmetic is almost trivial here explore this in Python a little We will start with a function to create a Telescoping Series. We will include a print statement to have a look at the list of values:  

```python
def compute_telescoping_series_with_arbitrary_power(power_choice,
                                                    upper_bound_choice):
    list_of_values = [i^power_choice - (i - 1)^power_choice for i in range(1, upper_bound_choice + 1)]
    print("---List of values:---\n", list_of_values)
    print("---Sumed list of values:---\n", sum(list_of_values))
    print("---Compute", str(upper_bound_choice), "^", str(power_choice), ":---\n", upper_bound_choice^power_choice)
    return(sum(list_of_values))

```

We note that we can pass it a value for a power, and an upper bound. It uses the telescoping series formula to find an aswer, but it also uses the values to formulate $u^p$. Let's try this function with a few examples: 

```python
compute_telescoping_series_with_arbitrary_power(2,10)
```

```python
compute_telescoping_series_with_arbitrary_power(8,20)
```

This behaves as we would expect. Now let's return to the binomial theorem. Recall for the form $(a - b)^k$ it would be: 



$$ (a + b)^k = \sum_{k = 0}^n \binom{n}{k}a^{n - k}b^k$$

Let's create a function for this: 

```python
def compute_binomial_theorem(a, b, n):
    list_of_values = [sp.binomial(n, k) * a^(n - k) * b^k for k in range(n + 1)]
    return(sum(list_of_values))
    
```

It behaves like we expect. Let's trysome 

```python
a, b = sp.symbols('a b')
s = compute_binomial_theorem(a, b, 4)
s
```
