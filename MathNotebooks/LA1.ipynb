<b>Observe</b> that there are number of different types of geometries making different assumptions regarding parralellism and perpendicularity (including Euclidean, Projective, Relativistic, Hyperbolic etc.)

<b>Observe</b> the type of geometry that makes the least underlying assumptions is Affine geometry, and is well suited to Linear Algebra can be visualised, and 2-dim affine geometry is the core problem of Linear Algebra


<b>Definition</b>: <b>Affine Geometry</b> is a type of geometry that makes following assumptions: 
- $A1$: A straight edge can be used to construct lines. <br/>
- $A2$: A straight edge can be construct lines that are parallel to other lines<br/>
- $A3$: Points can be constructued<br/>
- $A4$: Lines can be constructed through points<br/>
- $A5$: Counting across lines and points is linear numbers in operations of addition, subtraction, multiplication and division but you are limited to rational numbers. <br/>
