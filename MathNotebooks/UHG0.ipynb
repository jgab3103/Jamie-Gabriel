#### Lecture 1: Overview of Universal Hyperbolic Geometry

The aim of this lecture series is to provide a far simpler approach to hyperbolic geometry, which will be called <b>Universal Hyperbolic Geometry</b>. It will diverge from standard treatments and cover the following topics: 

- Appollonius and Conics
- Polarity
- Cross Ratio
- Projective Geometry
- The theorems of Boyai, Lobachesky, Gauss, Beltrami, Einstien and Minkowski
- Rational Trigonometry

Hyperbolic Geometry starts with different assumptions to Euclidean Geometry, as will this lecture series. In terms of other treatments of Hyperbolic Geometry, the difference in this lecture series will be not to employ the standard starting models (such as the Poincare Disk Model, or the half plane Betrami Klien Projective model). This treatment will also consider the properties of lines inside and outside the disc.  

The intent of this lecture series treatment allows a simplification of standard approaches to the subject, along with the possibility of avoiding trascendental functions, square roots, etc. Its formulas are more general and simpler. 

#### Lecture 1: Appolonius and Conics

The lectures start with the work of Apollonius of Perga (260-190 B.C) 

Appollonius undertook a study of conics and provided many propositions on the nature of conics. A a guiding intuitioin, the study of conics is the study of a cone structure (seen below, called a conic) and the questions and properties that can arise when a plane is intersected with this structure. This intersection of plane and conic will create geometric shapes such as the ellipse, hyperbola, parabola or circle. An example of this in action can be seen in Figure 1.

Note that the Greeks had been studying conics prior to Appolonius. 

The key insight from Appolonius was the understanding that the circle provides a duality between lines and points. 

Consider the construction in the image below: 

* Let $A$ be a point and construct a circle with this point as its centre
* Let $C$ be a point outside the circle
* Let $CD$ be a line that passes through the point $C$ and is tangent to the circle at point $D$
* Let $CE$ be a line that passes through the point $C$ and is tangent to the circle at point $E$
* Let $DE$ be a line that passes through the points $D$ and $E$

```python
Image("./Images/UHG_2.PNG", width = 400)
```

Note that by definition: 

* The line $DE$ is the polar of the point $C$
* The point $C$ is the pole of the line $DE$

This relationship of a point and line that could be constructed with any point outside the circle and two tangents is a relationship between pole and and polar. It is denoted $ DE = C^\perp $ or $C = DE^\perp $ where $\perp$ can be pronounced "perp"

The parlance here is that <i>"C equals DE perp"</i> or <i>"DE equals C perp"</i>

#### Projective Definition of Polarity

There is an alternate definition of Polarity that can be used in which, only a straight edge for a given circle. Note that UHG assumes a circle is always given for any construction, and it will be denoted as the <b>Null Circle</b>

In order to create this construction: 


* Let $C$ be any point
* Let $CD$ be a line that passes through $C$ and $D$
* Let $CE$ be a line that passes through $C$ and $E$
* Join all points that have been created on the circle ($FI, IH, HG, GF, GI, FH$)
* Note the interesection of these lines at $A$ and $K$ and create a line that passes through $A$ and $K$
* Note that the line $AK$ is the polar of $C$
* Note that $AK$ passes through the circle in 2 points, $L$ and $M$. These can be used to construct tangents from the point $C$

Note that this construction provicdes a definition of the tangent, but not require the tangent. 

<b>Polar independence theorem</b>

The polarity of a point and a circle does not depend on the choice of lines through the null circle. 

Note that this is just the construction above. The advantage of this definition of polarity is that it only requires a straight edge and the point casn be constructed inside or outside the null circle.
