<b>Aim</b>: Extend the summative propery of orientated edges to arbitrarily sized polygons to determine thier area.

<hr/>

<b>Observe</b>: It is possible to extend the notion of area as being the summation of the orientated sides of a triangle to polygons of an arbitrary number of sides.

<b>Observe</b>: As part the generalisation of this formula, a new definition of one of the components of the signed area of an orientated triangle, $s(\widetilde{OAB} \equiv \frac{1}{2}(ad - bc)$ will be introduced, called the cross product. 

<b>Defintion</b>: The Cross Product for 2 points $A = [a, b]$ and $B = [c, d]$ is defined as $ac - bd$

<b>Observe</b>: The implication of introducing the cross product is that the signed area of an orientated triangle ($\widetilde{OAB}$) or orientated edge ($\vec{AB}$) (where $O, A, B$ are points), is equivalent to half its cross product. 

<b>Definition</b>: and Orientated Polygonal Spline (OPS) $\overrightarrow{A_0,A_1\ldots,A_n} = [A_0,A_1\ldots,A_n]$ is an ordered list of affine points, $A_0, A_1, \ldots A_n$. 

<b>Theorem</b>: The signed area of an orientated polygonal spline is defined to be: 

$$ s(\overrightarrow{A_0,A_1\ldots,A_n}) = \Sigma_{k = 0}^{n-1} s(\overrightarrow{A_kA_{k + 1}})$$

<b>Observe</b>: If this If $A_0 = A_n$, then $s(\overrightarrow{A_0,A_1\ldots,A_n}) $ is a closed polygon and independent of translation. 

<b>Theorem (Meister's Formula)</b> Meister's formula calculates the signed area for an arbitrary number of points. In the case of 4 points $A = [x_1, y_1], B = [x_12, y_2], C = [x_3, y_3]$ and $D = [x_4, y_4]$, Meister's formula can be computed as: 


$$ s(\widetilde{A_1A_2A_3A_4})=s(\widetilde{A_1A_2})+s(\widetilde{A_2A_3})+s(\widetilde{A_3A_4}+s(\widetilde{A_4A_1})\\=\frac12(x_1y_2-y_1x_2)+\frac12(x_2y_3-y_2x_3)+\frac12(x_3y_4-y_3x_4)+\frac12(x_4y_1-y_4x_1))\\=\frac{1}{2}(x_1y_2-y_1x_2+x_2y_3-y_2x_3+x_3y_4-y_3x_4+x_4y_1-y_4x_1).  $$


<b>Observe</b>: Meister's formula follows from the additive property of the area of orientated edges. 

<b>Let</b> $X3, X4$ and $X5$ be function that implement the cross product, the signed area of an orientated edge, and the signed area using Meister's formula.
