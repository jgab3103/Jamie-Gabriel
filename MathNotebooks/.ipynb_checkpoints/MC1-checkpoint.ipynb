Last time we explored Projective geometry. Let's change the rules, and Affine Geometry. The rules of Affine Geometry are the following:

- You are given a straight edge so you can construct lines
- You can construct parallel line to a line
- You can see where lines meet. 
- Just because you cannot measure things, you can still count things

Note this is mroe restrictive than the last geometry we encountered. There is no notion of distance, no special point that denotes an origin. There are no angles, there are no curves we can use. THere is no perpendicularity 

Eventually we will view the different types of geometries (such as Cartesian, Hyperbolic, Euclidean) as different families. But Affine is really the simpest, and those other geometries are supersets of the geometries. Affine Geometry allows us to be as general as posssible is one of the simpest ways we can interact with space.
