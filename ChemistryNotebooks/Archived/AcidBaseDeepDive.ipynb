- Acids usually have H in front
     - HCl - hydrogen with non metal
     - HF
     - HC2H3O2
     - Has positive charge

- Bases usually have hydroxide ion
    - NaOH

Arrhenius Definion

- Acids -> release H+ ions into solution, so H+ becomes H3O+ in water
- Bases -> releases OH- into solution


Bronston-Lowrey Definiion

- Acides are proton donors
- Bases are proton acceptors

-EG: HCL (acid) + H2O (base) -> Cl-(conjugate base, take away ion) + H3O+ (conjugate acid, gets ion)

- Note that Acid -> Conjugate Base
- Note that Base -> Conjugate Acid


- Eg. NH3(Base) + H2O(Acting as acid) -> NH4+ (Conjutage acid, acceptor) + OH-(Conjugate base, donor)




What is conjugate acid and conjugate base of water

- H2O -> H3O+    :: Becomes conjugate acid, acceptor, gains H
- H2O -> OH-     :: Becomes conjugate acid, loses H 


Eg. Write conjugate acid and conjugate base of NH3, Ammonia. 

- NH3 -> NH4+  :: gains H, is conjugate acid
- NH3 -> NH2-  :: loses H, is conjugate base

Eg. Write conjugate acid and conjugate base of H2PO4-

- H2PO4- -> H3PO4
- H2PO4- -> HPO4-2

- Less than 7: Acidic
- Greater than 7: Basic

##### Strong Acids and Weak Acids

- Strong acids ionise completely (100%)
- Weak acides partially ionise (<5%)

- Strong Acides for strong electolytes, conduct electricity

- Strong Acids are: 
    - HCl
    - HBr
    - HI
    - HNO3
    - H2SO4
    - HClO4
  
  
- Weak Acids
    - HF
    - NH4+
    - HC2H3O2
    - HNO2
    - H2SO4
    - HNC
    - anything that is not on strong list
